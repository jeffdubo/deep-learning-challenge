## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(columns=['EIN', 'NAME'], inplace=True)
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = [i for i in application_type_counts.index if application_type_counts[i] < 500]

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts[classification_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Code Ref: https://stackoverflow.com/questions/16476924/how-to-iterate-over-rows-in-a-dataframe-in-pandas
classifications_to_replace = [i for i in classification_counts.index if classification_counts[i] < 1000]

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df.drop(columns='IS_SUCCESSFUL')
y = application_df['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

# Display value counts for y_train and y_test to check distribution of values
print('y_train:')
print(y_train.value_counts())
print('\ny_test:')
print(y_test.value_counts())
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

# Create function to define model and compile to test different parameters
def create_model(num_input_features, nodes_l1, nodes_l2, nodes_l3, activation_l1, activation, optimizer):

    # Clear old sessions/models
    # Code Ref: https://www.tensorflow.org/api_docs/python/tf/keras/backend/clear_session
    tf.keras.backend.clear_session()
    nn = tf.keras.models.Sequential()

    # Hidden Layers
    nn.add(tf.keras.layers.Dense(units=nodes_l1, input_dim=num_input_features, activation=activation_l1))
    nn.add(tf.keras.layers.Dense(units=nodes_l2, activation=activation))
    # Add 3rd layer if nodes_l3 > 0
    if nodes_l3 > 0:
        nn.add(tf.keras.layers.Dense(units=nodes_l3, activation=activation))
    # Output layer
    nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Check the structure of the model
    display(nn.summary())

    # Compile the model
    nn.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    return nn
```

```python
# Create and train Model #1
#   First Layer: 80 nodes, relu activation
#   Second Layer: 30 nodes, relu activation
#   Output Layer: adam optimizer
num_input_features = len(X_train.iloc[0])
nn = create_model(num_input_features, 80, 30, 0, 'relu', 'relu', 'adam')
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate Model #1
#   First Layer: 80 nodes, relu activation
#   Second Layer: 30 nodes, relu activation
#   Output Layer: adam optimizer
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

### Analysis: Model #1 did not perform well with an accuracy of only 0.725 and loss of .556.

## Create, Train, and Evaluate Additional Models using Keras Tuner

```python
# Create a function that creates a new Sequential model with hyperparameter options
def create_model_hp(hp):

    tf.keras.backend.clear_session()
    nn = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh'])

    # Allow kerastuner to decide number of neurons in first layer
    nn.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=50,
        max_value=100,
        step=10), activation=activation, input_dim=num_input_features))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 5)):
        nn.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=10,
            max_value=50,
            step=10),
            activation=activation))

    nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn.compile(loss='binary_crossentropy', optimizer='Nadam', metrics=['accuracy'])

    return nn
```

```python
# Import the kerastuner library
import keras_tuner as kt

# Initiate tuner
tuner = kt.Hyperband(
    create_model_hp,
    objective='val_accuracy',
    max_epochs=30,
    hyperband_iterations=2)

tuner.search(X_train_scaled, y_train, epochs=20, validation_data=(X_test_scaled, y_test))
```

```python
# Display evaluation and hyperparamters for top 3 models
for i in list(range(0, 3)):
    # Get model and hyperparameters
    model = tuner.get_best_models(3)[i]
    hyper = tuner.get_best_hyperparameters(3)[i]

    # Evaluate model
    model_loss, model_accuracy = model.evaluate(X_test_scaled,y_test,verbose=2)

    # Print model evaluation and hyperparamaters
    print(f'Model Rated #{i+1} Evaluation:')
    print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
    print(f'Model Rated #{i+1} Hyperparamaters:')
    display(hyper.values)
```
